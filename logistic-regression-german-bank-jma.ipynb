```r
german_bank <- read.csv("../input/german-bank/german_bank.csv",header=T)
#Explanation: We are importing the data set german bank and using them for logistic regression.
str(german_bank)
#Explanation : The numerical variables are treated as integers over here.
german_bank <- german_bank[,2:32]
#Explanation : We are selecting all the variables from 2 to 32 because the first variable i.e observation number are not used for our analysis.
for (i in c(1,3:9,11:21,23:31)){ german_bank[,i] <- factor(x=german_bank[,i],levels=sort(unique(german_bank[,i]))) 
}
#Explanation : The required variables are transformed into factors from the german_bank data set.
class(german_bank)
#Explanation : The german bank is a data set.
set.seed(25)
#Explanation : We are generating random numbers using set.seed option.
ranuni <- sample(x=c("Training","Testing"),size=nrow(german_bank),replace=T, prob=c(0.7,0.3))
ranuni
TrngDt <- german_bank[ranuni == "Training",]
TstDt <- german_bank[ranuni == "Testing",]
#Explanation: The data sets are divided into trainingdata and testingdata ,i.e 70% of the observations are in the TrngDt and the rest 30% of the observations are stored in TstDt.

indVariables <- colnames(german_bank[,1:30])
#Explanation: the Independent Variables are selected from 1 to 30.


rhsOfModel <- paste(indVariables,collapse="+")
rhsOfModel
#Explanation: The independent variables are collapsed with a additional operator.
model=paste("RESPONSE~",rhsOfModel)
model
#Explanation: The dependent variable is equated to the independent variables. Here ~ sign is used in place of the equal to sign in an equation which is known as the relational operator. All these steps are done in order to build the model.
frml <- as.formula(model)
frml
#Explanation: The model is changed to a formula data type using the as.formula function.
#BUILDING the Training Model
TrainingModel <- glm(formula=frml,data=TrngDt,family="binomial")
TrainingModel
summary(TrainingModel)
class(TrainingModel)
TrainingModel <- step(object=TrainingModel,direction="both")
summary(TrainingModel)
#Explanation: Here forward and backward regression is done with the help of the TrainingModel. Here the model with minimum AIC values are taken as the optimum model . From the summary function we get the p values and the 6 statistical measures of the residuals terms.

library(pROC) 
#Explanation: This library is installed and activated  for the further analysis. 
troc <- roc(response=TrainingModel$y,predictor=TrainingModel$fitted.values,plot=T)
troc$auc

TrPred <- ifelse(test=TrainingModel$fitted.values < 0.5, yes= 0, no= 1)
table(TrainingModel$y,TrPred)
#Explanation : Here we are doing the classification table for the training model with the help of the training data, 
#i.e. a table with the observed and the predicted probabilities of event(1) and non event that is zero(0) . With the help of this table we calculate the sensitivity, specificity, percentage correct, false positive and false negative. The value of the percentage correct shows how much of the model is correctly classified. Higher is the value of the percentage correct better is the model. The percentage classified for the above model is 79.15%.
library(pROC) 
#Explanation: This library is installed and activated  for the further analysis.
tsPred <- predict.glm(object=TrainingModel,newdata=TstDt,type="response")
#Explanation: We are predicting the values of the dependent variable using the testing data. Here 30% of the data are predicted.
tsPred
tsroc <- roc(response= TstDt$RESPONSE,predictor= tsPred,plot=T)
tsroc$auc
#Explanation: We are plotting the ROC curve (Receiver Operating Characteristic Curve) of our training model using the testing data and also calculating the area under ROC of the model which is very high that is it is close to 1 showing that the model is a good model.
#OUTPUT
#ROC CURVE FOR TESTING DATA
library(pROC) 
#Explanation : This library is installed and activated  for the further analysis.
trapred<- predict.glm(object=TrainingModel,data=TrngDt,type="response")
View(trapred)
#Explanation : Here based on the training model using training data 70% of the observations for the dependent variables are predicted.
#OUTPUT
tsPred<- ifelse(test=tsPred< 0.5, yes= 0, no= 1)
table(TstDt$RESPONSE,tsPred)
#Explanation : Here we are doing the classification table for the training model with the help of the testing data, i.e. a table with the observed and the predicted probabilities of event(1) and non event that is zero(0) . With the help of this table we calculate the sensitivity, specificity, percentage correct, false positive and false negative. The value of the percentage correct shows how much of the model is correctly classified. Higher is the value of the percentage correct better is the model. The amount of percentage classified of the model using the testing data is 74.522%.
#OUTPUT
```
