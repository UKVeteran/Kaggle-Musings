```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../input/restaurant-reviews/Restaurant_Reviews.tsv', delimiter='\t', quoting=3)
df.head()
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

def rep(review):
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    reviews = review.split()
    reviews = [stemmer.stem(x) for x in reviews if x not in stopwords.words('english')]
    review = ' '.join(reviews)
    return review
```

```python
df['Review'] = df['Review'].apply(rep)
df.head()
```

```python
corpus = df['Review'].tolist()
corpus[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=1500)
X = cv.fit_transform(corpus).toarray()
y = df['Liked'].tolist()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
clf.fit(X_train, y_train)
print ('Accuracy:{0: .1f}%'.format(clf.score(X_test, y_test) * 100))
```
